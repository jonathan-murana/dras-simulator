# client

```python
ax = plt.figure().add_subplot(111)
ax.set_xlabel('offer (DOLAR) ')
ax.set_ylabel('reduction (WATTS)')

for filename in [f for f in os.listdir(report1_dir +'/') if "out_client" in f]:
    data = pd.read_csv(report1_dir +'/' +filename, header=None,sep=' ',skiprows=0,)
    data.columns = [
        'price', 'alpha_' + filename,'profit', 'loss','payment', 'reductionPercentage','nonCompleteTasks']
   
    ax = data[['alpha_' + filename]].plot(ax=ax)
    print(filename + '=' + str(data['alpha_' + filename].nunique())) 
```
